```python
import umap, umap.plot
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
from scipy.interpolate import CubicSpline
```

```python
data = np.loadtxt("../../joint.csv", delimiter=",")
len(data)
```

```python
S1 = [ 0.2, 0.2, 0.5, 0.4, -0.4, -0.1, -0.5, -0.5 ]
S2 = [ -0.6, 0.6, 0.1, 0.0, 0.0, 0.05, 0.25, 0.25 ]
S3 = [ -0.4, 0.4, 0.1, -0.2, -0.2, -0.5, -0.5, -0.5 ]
S4 = [ 0.0, 0.0, -0.05, 0.6, -0.6, -0.1, -0.25, -0.25 ]
S5 = [ 0.2, 0.2, 0.5, 0.4, -0.4, -0.1, -0.5, -0.5 ]
waypoints = np.vstack([S1, S2, S3, S4, S5]) 
t_nodes = np.array([0.0, 0.2, 0.5, 0.8, 1.0])

spline = CubicSpline(t_nodes, waypoints, bc_type='periodic', axis=0)
```

```python
dataarr = []
for i in range(10000):
    dataarr.append(spline(0.0001*i))
```

```python
# Fit UMAP to 3D
mapper = umap.UMAP(n_components=3).fit(data)
embedding = mapper.embedding_

# Create color values (time index here)
colors = np.arange(data.shape[0])

# Interactive scatter plot
fig = px.scatter_3d(
    x=embedding[:, 0],
    y=embedding[:, 1],
    z=embedding[:, 2],
    color=colors,
    color_continuous_scale='Viridis',
    title='3D UMAP Embedding',
    opacity=0.7
)

fig.update_traces(marker=dict(size=3))
fig.show()
```
